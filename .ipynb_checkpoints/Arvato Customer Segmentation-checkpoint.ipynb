 > | Column Name |  Non null rows | No Value Preset | Percentage Null | Coloumn Desciption

```python
def get_nullpercentage(df, desc_df):
    # Iterate through column and create details
    col_details = []
    n_rows = df.shape[0]

    for index, col in enumerate(df.columns):
        null_count = df[col].isnull().sum()
        percentage_nulls = math.floor(((null_count/n_rows) * 100) * 10)/10
        non_nulls = n_rows - null_count

        col_details.append((col, non_nulls, null_count, percentage_nulls))

    # Create Dataframe and from list

    col_details_df = pd.DataFrame(col_details, columns=['Column Name', 'Non Null Rows', 'Null Rows', 'Percentage Null'])

    # Use the **azdisa_attrs** Data frame in this function 'desc_df' code cell 3 to get the column Information and concat to col_details_df
    
    col_description = desc_df['Description']
    col_details_df = pd.concat([col_details_df, pd.Series(col_description)], axis=1)
    col_details_df = col_details_df.sort_values(by='Percentage Null', ascending=False)

    return col_details_df

def plot_nulls(df, x, y,title, threshold=False, limit=None):
    '''
    Description: Plot a bar chart from the given points
    
    Parameters:
        df : Dataframe
        x : label or position
        y : label, position or list of label, positions, default None
        title: Title of the Plot, default to 'x agains y'
        threshold: boolean, defauls none
        limit: Number of points
    '''
   
    df = df[:limit]
    
    df.plot(kind='bar', 
                x=x, 
                y=y,
                title = title,
                figsize=(20,10), 
                fontsize=14,
                sort_columns=True,
                mark_right=True);
```

```python
def topology_float_to_int_type_con(string):
    '''
    Description: Convert float to interget and then to string
    
    Parameters: 
        string: String to convert
    '''
    
    converted_float =  str(round(int(float(string)))) 
    return converted_float

def clean_data(df, droprows=True, scale=True):
    '''
    Desciption:
        Takes a data frame cleans it and produced a cleaned dataframe
        
    Parameters:
        df : Dataframe
        droprows: Drop null rows or not, defauls to True
        scale: Scale resulting dataframe or not, defaults to True
    
    Output:
        cleaned_df: Cleand Data Frame
    '''
    print('Cleaning DataFrame')
    print('Drop any column with more than 60% Missen Values')
    df = df.drop(df.columns[df.isnull().sum()/df.shape[0] > .6], axis=1)
    print('Done')

    if droprows:
        # Threshold her is 70% meaning we require atleast 70 of the fields coloumns for a row to be null to delete.
        print('Drop any row with more than 70% Missen Values')
        df = df.dropna(thresh=math.floor(70/100 * df.shape[1]), axis=0)

    # Generate Corrolation Matrix
    df_corrmatix = df.corr()
    print('Calculating Corrolation')

    print('Removing Corolation matrix')
    # Removing Highly corrolated Matric take the absolute value
    azdias_corrmatix = df_corrmatix.abs()

    # Create a True/False mask and apply it
    mask = np.triu(np.ones_like(azdias_corrmatix, dtype=bool))

    tri_df = azdias_corrmatix.mask(mask)

    print('Drop Columns with corolation > .97')
    # List column names of highly correlated features (r > 0.97)
    to_drop = [c for c in tri_df.columns if any(tri_df[c] >  0.97)]

    # Drop the features in the to_drop list
    reduced_df = df.drop(to_drop, axis=1)

    print("Data Frame Columns Before and after removing corrolated columns")
    print("Before: {} ".format(df.shape[1]))
    print("After: {} ".format(reduced_df.shape[1]))


    # Data Imputation 
    
    categorical_cols = reduced_df.select_dtypes(include=['object'])
    category_names = categorical_cols.columns
    
    # Drope Categorical columns from oraginal dataframe 
    reduced_df = reduced_df.drop(columns=category_names, axis=1)
    
    # Fill Nan variable with  XX for CAMEO_DEU_2015 and -1 for CAMEO_DEUG_2015 and CAMEO_INTL_2015 columns 
    values = {'CAMEO_DEU_2015':'XX', 'CAMEO_DEUG_2015': '-1', 'CAMEO_INTL_2015': '-1'}
    categorical_cols = categorical_cols.fillna(value=values)

    # Fill X values of variable CAMEO_DEUG_2015 with -1 with represents the unknows values
    categorical_cols.loc[categorical_cols['CAMEO_DEUG_2015'] == 'X'] = '-1'


    # Fill XX values of variable CAMEO_INTL_2015 with -1 with represents the unknows values
    categorical_cols.loc[categorical_cols['CAMEO_INTL_2015'] == 'XX'] = '-1'

    # Perform forward and back filling on column D19_LETZTER_KAUF_BRANCHE
    categorical_cols['D19_LETZTER_KAUF_BRANCHE'] = categorical_cols['D19_LETZTER_KAUF_BRANCHE'].fillna(method='ffill').fillna(method='bfill')

    # Check for any XX values in CAMEO_DEUG_2015 categorical fields
    categorical_cols[categorical_cols['CAMEO_DEUG_2015'] == 'X']

    # Convert all string variable with numeric value of float to int for 'CAMEO_DEUG_2015' and 'CAMEO_INTL_2015'
    col_list = ['CAMEO_DEUG_2015', 'CAMEO_INTL_2015']

    for col in col_list:
        print('Doing String conversion for {}'.format(col))
        categorical_cols[col] = categorical_cols[col].apply(topology_float_to_int_type_con)
    print('All Float and Int equvalent whole numbers converted to string ')

    # Encounterd a problem of having strings of different formats long, short and missen string encoded as -1 solution from here
    # https://stackoverflow.com/questions/55440145/pd-to-datetime-valueerror-given-date-string-not-likely-a-datetime

    ## I will drop this coloumns 
#     date_long = pd.to_datetime(categorical_cols['EINGEFUEGT_AM'],format='%Y-%m-%d %H:%M:%S', errors='coerce')
#     date_short = pd.to_datetime(categorical_cols['EINGEFUEGT_AM'],format='%Y-%m-%d', errors='coerce')

#     # Combine the short dates to long dates
#     categorical_cols['EINGEFUEGT_AM']= date_long.combine_first(date_short)


#     categorical_cols['EINGEFUEGT_AM_Year'] = categorical_cols['EINGEFUEGT_AM'].dt.year
#     categorical_cols['EINGEFUEGT_AM_Month'] = categorical_cols['EINGEFUEGT_AM'].dt.month
#     categorical_cols['EINGEFUEGT_AM_Day'] = categorical_cols['EINGEFUEGT_AM'].dt.day
    
#     # Drope EINGEFUEGT_AM column as its values have already been captured above into year, month and day
    
    categorical_cols = categorical_cols.drop(columns='EINGEFUEGT_AM', axis=1)
    
    # Fill Nan variable with  XX for CAMEO_DEU_2015 and -1 for CAMEO_DEUG_2015 and CAMEO_INTL_2015 columns 
    
#     values = {'EINGEFUEGT_AM_Year':'-1', 'EINGEFUEGT_AM_Month': '-1', 'EINGEFUEGT_AM_Day': '-1'}
#     categorical_cols = categorical_cols.fillna(value=values)
    
#     categorical_cols = categorical_cols.drop(columns=['EINGEFUEGT_AM_Year', 'EINGEFUEGT_AM_Month', 'EINGEFUEGT_AM_Day'], axis=1)
   
#     categorical_cols = categorical_cols.drop(columns=['EINGEFUEGT_AM_Year'], axis=1)

    print("Replacing Nan with mean for numeirc variables")
    # Use sklearn's Imputer class on oraginal Dataframe to do imputation of missind Nan Values for numric colums using the Column mean
    imp = Imputer(missing_values=np.nan, strategy='mean')
    reduced_matrix = imp.fit_transform(reduced_df)
    reduced_df = pd.DataFrame(reduced_matrix, columns=reduced_df.columns)
    
    categorical_cols = pd.get_dummies(categorical_cols) 
    
    reduced_df = reduced_df.reset_index(drop=True)
#     categorical_cols = categorical_cols.reset_index(drop=True)
    
    for col in categorical_cols.columns:
        reduced_df[col] = categorical_cols[col].reset_index(drop=True)
    
#     cleaned_df = pd.concat([reduced_df, categorical_cols], axis=1)

    if scale:
        print('Scaling DataFrame')
        scaler = StandardScaler()
        scaled_df = scaler.fit_transform(reduced_df.astype('float'))
        azdias_scaled = pd.DataFrame(scaled_df, index=reduced_df.index, columns=reduced_df.columns)
        reduced_df = azdias_scaled
        
        print('Scaling Complete')
    
    print('Cleaning Complete')
    
    reduced_df = reduced_df.drop(columns=['LNR'], axis=1)
    
    return reduced_df

```

### Reducing Dimmensionality 
Here we seek to reduce the variance among features of the dataset by reducing the dimensions of our dataset, this avoids the curse of dimmensionality, when the number of dimmensions in the dataset is huge or greatly inceased it leads to data sparcity. Also this also reduces the possiblity of overfitting due to adding more feartures that extend our model in new data. 

https://en.wikipedia.org/wiki/Principal_component_analysis

For dimmensional reducation we use the PCA in sklearn which uses an orthogonal transformation to convert a set of observations of possibly correlated variables,yielding an orthogonal basis in which different individual dimensions of the data are uncorrelated The standardisation above helps make this process efficient as PCA is very sensitive to scale.
These basis vectors are called principal components

This resulting vector will be a vector of maximum variance

### K Mean Clustering. 
Presently our data has alot of dimmension which is difficult to see relationships between rows, we user an Unsupervised learning process here to cluster similar data points based on the feature correlation between themset into clusters with smaller data items.

We user KMean will be used for clustering
First we user the elbow point to determin an approprate number of clusters.

Finall We compare clusters for Azdias and those of customers to see pottential customers

Steps;
    - Determine optimal clusters by try for 25 diffent cluster numbers (Elbo Method)
    - Plot Number of clusters agains score
    - Determin number of clusters and generat final clusters 
    - Plot a scatter map to show differnt clusters

```python
%%time
def plot_kmean_distortion(df, n): 
    # We can easily run K-Means for a range of clusters using a for loop and collecting the distortions into a list

    distortions = []
    K = range(1, n)
    for k in K:
        print('Fitting for cluster {}'.format(k))
        kmeanModel = KMeans(n_clusters=k)
        kmeanModel.fit(df)
        distortions.append(kmeanModel.inertia_)

    plt.figure(figsize=(16,8))
    plt.plot(K, distortions, 'bx-')
    plt.xlabel('k')
    plt.ylabel('Distortion')
    plt.title('The Elbow Method showing the optimal k')
    plt.show()
    
plot_kmean_distortion(reduce_azdias_df, 25)
```

From the distortion above we can see that the Elbow point is betwenn 10 and 15  clusters.

A Receiver Operator Characteristic (ROC) curve is a graphical plot used to show the diagnostic ability of binary classifiers. 0.5 suggests no discrimination (i.e., the ability of not producing false positives ), 0.7 to 0.8 is considered acceptable, 0.8 to 0.9 is considered excellent, and more than 0.9 is considered outstanding

```python
def train_predict_score(estemators, X_train, X_test, y_train, y_test, with_grid_cv=False, param_grid ={}):
    '''
    Description:
        Use Estemator to train a model run predictions, display fscore and other details
        
    Parameters:
        estemators: An Estimator
        X_train: Data to Fit to Model
        y_train: Target Value
        X_test: Data test
        y_test: Test Data
        with_grid_cv: If to apply grid search or not, Defaults(False)
        param_grid: Parameters to Grid Search defaults to an exmpty object
        
    Result:
        mode_df: DataFrame of Details of the training process
        regr_model: Built Models in a dictionry
    '''
        
    model_res = []
    mode_df = []
    regr = []
    regr_model = dict()
    
    for estem in estemators:
        print('{} Algorithm starting'.format(estem.__class__.__name__))
        
        if with_grid_cv:  
            regr = GridSearchCV(estimator=estem, param_grid=param_grid , scoring='roc_auc', cv=5)
            regr.fit(X_train, y_train)
            y_pred =  regr.predict(X_test)
            
            regr_rscore = accuracy_score(y_test, y_pred) 
            
            square_error = mean_squared_error(y_test, y_pred) #The mean squared error tells you how close a regression line is to a set of points

            print(regr.best_estimator_)
            
        else :     
            regr = copy.deepcopy(estem)
            regr.fit(X_train, y_train)
            y_pred = regr.predict(X_test)
            
            regr_rscore = accuracy_score(y_test, y_pred)
            square_error = mean_squared_error(y_test, y_pred) #The mean squared error tells you how close a regre
        
        # Put Informinto a dictionary
        regr_dic = dict()
        model_name = estem.__class__.__name__
        
        regr_dic['Algorithm'] = model_name
        regr_dic['Accuracy'] = regr_rscore 
        regr_dic['Percentage Accuracy'] = round(regr_rscore * 100, 2)
        regr_dic['Square Error'] = round(square_error, 4 )

        if with_grid_cv:
            regr_dic['best_score'] = regr.best_score_
            regr_model[model_name] = regr.best_estimator_
        else:
            regr_model[model_name] = regr
            
        model_res.append(regr_dic)
        
    
    model_res = sorted(model_res, key=lambda k: k['Accuracy'], reverse=True ) #Sort list in rscore values 
    
    mode_res_df = pd.DataFrame.from_dict(model_res)

    
    # Return the model alongside the data frame.
        
    return mode_res_df , regr_model
```

```python
def plot_feature_importance(n, fi, labels):
    '''
    Description: Plot of feature importance 
    parameters: 
        n: Number of features to display
        fi: Feature Importance
        labels: List of columns names
    '''
    feature_importance = fi[:n]
    sorted_idx = np.argsort(feature_importance)
    pos = np.arange(sorted_idx.shape[0]) + .5
    fig = plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.barh(pos, feature_importance[sorted_idx], align='center')
    plt.yticks(pos, np.array(labels[:n])[sorted_idx])
    plt.title('Feature Importance (MDI)')
    fig.tight_layout()
    plt.show()
    
plot_feature_importance(10,gregr_model['GradientBoostingClassifier'].feature_importances_, cleaned_mailout.columns )
```
